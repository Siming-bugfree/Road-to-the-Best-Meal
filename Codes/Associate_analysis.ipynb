```python
import pandas as pd
token_recipe = pd.read_csv('PP_recipes.csv')
```

```python
token_recipe
```

```python
from apyori import apriori
import matplotlib.pyplot as plt
```

```python
import json
ing_list = []
for i in range(0, 178265):
    ing_list.append(json.loads(token_recipe['ingredient_ids'][i]))
    
association_rules = apriori(ing_list, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
for item in association_results:

    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + str(items[0]) + " -> " + str(items[1]))

    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
import pickle
with open('ingr_map.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data
```

```python
data = data[['replaced', 'id']].drop_duplicates()
```

```python
for item in association_results:

    pair = item[0] 
    items = [x for x in pair]
    A = data[data['id'] == items[0]].iloc[0]['replaced']
    B = data[data['id'] == items[1]].iloc[0]['replaced']
    if B == 'all-purpose flmy':
        B = 'all-purpose flour'
    print("Rule: " + A + " -> " + B)

    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
def bubble_sort(nums):
    # We set swapped to True so the loop looks runs at least once
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i][2][0][3] < nums[i + 1][2][0][3]:
                # Swap the elements
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                # Set the flag to True so we'll loop again
                swapped = True
```

```python
top_ten = association_results[:10]
```

```python
bubble_sort(top_ten)
```

```python
for item in top_ten:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    A = data[data['id'] == items[0]].iloc[0]['replaced']
    B = data[data['id'] == items[1]].iloc[0]['replaced']
    if B == 'all-purpose flmy':
        B = 'all-purpose flour'
    print("Rule: " + A + " -> " + B)

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
