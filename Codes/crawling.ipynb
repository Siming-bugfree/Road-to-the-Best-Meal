```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
import numpy as np 
import datetime
```

```python
df = pd.read_csv('data_for_recommendation.csv')
user_id = list(df['user_id'].unique())
```

```python
url = 'https://www.food.com/user/'
# follower_list = []
# following_list = []
# rate_list = []
# year_list = []
# month_list = []

for i in range(27396, len(user_id)):
    response = requests.get(url + str(user_id[i]))
    if response.status_code == 200:
        print("{} is Success".format(i))
        results_page = BeautifulSoup(response.content,'lxml')
        user = results_page.find_all('div', class_ = 'data')
        follower = int(user[0].text)
        following = int(user[1].text)
        try:
            rate = float(results_page.find_all('span', class_ = 'rating-text')[0].text.split(' ')[0])
        except:
            rate = np.nan
        try:
            year = datetime.datetime.strptime(results_page.find_all('div', class_ = 'user-facts')[0].text.strip().split(' ')[-1], '%m/%Y').year
            month = datetime.datetime.strptime(results_page.find_all('div', class_ = 'user-facts')[0].text.strip().split(' ')[-1], '%m/%Y').month
        except:
            year = np.nan
            month = np.nan

        follower_list.append(follower)
        following_list.append(following)
        rate_list.append(rate)
        year_list.append(year)
        month_list.append(month)
        
    else:
        print("{} is Failure".format(i))
        follower_list.append(np.nan)
        following_list.append(np.nan)
        rate_list.append(np.nan)  
        year_list.append(np.nan)  
        month_list.append(np.nan)  
```

```python
user_feature = pd.DataFrame(user_id).set_index(0)
user_feature['follower_num'] = follower_list
user_feature['following_num'] = following_list
user_feature['rate'] = rate_list
user_feature['join_year'] = year_list
user_feature['join_month'] = month_list
user_feature
```
