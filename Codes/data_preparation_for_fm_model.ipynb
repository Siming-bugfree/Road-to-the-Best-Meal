```python
inter_df = pd.read_csv('RAW_interactions.csv')
#inter_df.drop('review', axis=1, inplace=True)
def with_indexing(dstr):                              
    return datetime.date(*map(int, [dstr[:4], dstr[5:7], dstr[8:10]])) 
inter_df['date'] = inter_df['date'].apply(with_indexing)
inter_df
```

```python
#在recip加入平均ratings
recip = pd.read_csv('RAW_recipes.csv')
recip.drop('steps', axis=1, inplace=True)
recip.set_index('id',inplace=True)
recipe_rating = inter_df.groupby('recipe_id').mean()['rating']
recip = recip.merge(pd.DataFrame(data = {'recipe_rating':recipe_rating.values}, index=recipe_rating.index), left_index=True, right_index=True)
#recip.to_csv('recip.csv')
```

```python
#合并recip和nutrition
nutri_df = pd.read_csv('nutri_df.csv')
nutri_df = nutri_df.iloc[:,1:-1]
recip.reset_index(inplace = True)
recip.columns = ['recipe_id', 'name', 'minutes', 'contributor_id', 'submitted', 'tags',
       'nutrition', 'n_steps', 'description', 'ingredients', 'n_ingredients',
       'recipe_rating']
recip = pd.merge(recip, nutri_df, on = 'recipe_id')
```

```python
recip.set_index('recipe_id', inplace = True)
```

```python
#加入ingred被评价次数
recipe_interactions = inter_df.groupby('recipe_id')['user_id'].count()
recip = recip.merge(pd.DataFrame(data = {'recipe_interactions':recipe_interactions.values}, index=recipe_interactions.index), left_index=True, right_index=True)
```

```python
recip  = recip[['minutes',
       'n_steps', 'n_ingredients',
       'recipe_rating_x', 'calories', 'total_fat', 'sugar', 'sodium',
       'protein', 'saturated_fat', 'carbohydrates',
       'recipe_interactions']]
```

```python
#把菜谱的tag转化成特征
tag_info = pd.read_csv('tag_info.csv')
tag_info = tag_info.iloc[:,1:3]
tag_info = tag_info.groupby(['recipe_id','tags']).size().unstack()
tag_info = tag_info.fillna(0)
```

```python
#合并数据集
```

```python
recip = recip.merge(tag_info, left_index=True, right_index=True)
```

```python
#准备user特征
user = pd.read_csv('user_feature_scraw.csv')
```
